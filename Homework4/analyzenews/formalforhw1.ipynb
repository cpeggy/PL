```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#來源：中時新聞網
url1 = 'https://www.chinatimes.com/opinion/20221010003509-262110?chdtv'
r1 = requests.get(url1)
web_content1 = r1.text
soup1 = BeautifulSoup(web_content1,'lxml')
```

```python
articleContent1 = soup1.find_all('p')
articleContent1
```

```python
# 將所有tag p中的text取出並放進list。
article1 = []
for p in articleContent1:
    article1.append(p.text)
```

```python
articleAll1 = '\n'.join(article1)
print(articleAll1)
```

```python
import jieba
import nltk
```

```python
jieba.load_userdict('text.txt') # 輸入自己字典的路徑
```

```python
d1 = articleAll1.replace('[^\w\s]','').replace('／',"").replace('《','').replace('》','').replace('，','').replace('。','').replace('「','').replace('」','').replace('（','').replace('）','').replace('！','').replace('？','').replace('、','').replace('▲','').replace('…','').replace('：','').replace('※以上言論不代表旺中媒體集團立場※','').replace('中時新聞網對留言系統使用者發布的文字圖片或檔案保有片面修改或移除的權利當使用者使用本網站留言服務時表示已詳細閱讀並完全了解且同意配合下述規定','').replace('違反上述規定者中時新聞網有權刪除留言或者直接封鎖帳號請使用者在發言前務必先閱讀留言板規則謝謝配合','')
print(d1)
```

```python
jieba.setLogLevel(20)

Sentence = jieba.cut(d1, cut_all=True)
print('全模式'+": "  + "/ ".join(Sentence) + '\n')   
 
Sentence = jieba.cut(d1, cut_all=False)
print('精確模式'+": " + "/ ".join(Sentence)+ '\n')  
 
Sentence = jieba.cut(d1)  
print('Default為精確模式'+": " + "/ ".join(Sentence)+ '\n')
 
Sentence = jieba.cut_for_search(d1)  
print('搜索引擎模式'+": " + "/ ".join(Sentence)+ '\n')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 設定停用字詞 
stopwords = {}.fromkeys(["也","但","來","個","再","在","的","和","是","有","更","會","可能","有何","從","對","就", '\n','越','為','這種','多','越來',' ','停招','停辦','了','後','有些','或','讓','與','已','於','還是','堅持','1','2','3','4','5','6','7','8','9','0','他','她','個','十','百','千','萬','除了','之外','年','月','日','10','2022','然而','到', '上','下','左','右','左右','而','及','以外','會將','是否'])
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence1 = jieba.cut_for_search(d1) 
```

```python
# create a python dictionary
hash = {}
for item in Sentence1:

    if item in stopwords:
        continue
    
    if item in hash:
        hash[item] += 1
    else:
        hash[item] = 1
```

```python
# 文字雲樣式設定
wc = WordCloud(font_path="C:/Windows/Fonts/msjh.ttc", #設置字體
               background_color="white", #背景顏色
               max_words = 35 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
```

```python
# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)
```

```python
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(40,20), dpi =1600)
plt.show()
```
