```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#來源：奇摩新聞
url2 = 'https://tw.stock.yahoo.com/news/%E5%B0%91%E5%AD%90%E5%8C%96%E5%95%8F%E9%A1%8C%E5%9A%B4%E9%87%8D-%E7%A7%81%E6%A0%A1%E6%8E%80%E8%B5%B7%E9%80%80%E5%A0%B4%E6%B5%B7%E5%98%AF-%E4%BB%8A%E5%B9%B4%E5%88%86%E7%99%BC%E7%BC%BA%E9%A1%8D%E5%89%B5%E6%96%B0%E9%AB%98-51%E6%A0%A1%E6%8B%9B%E4%B8%8D%E6%BB%BF-%E9%80%80%E5%A0%B4%E6%A2%9D%E4%BE%8B%E9%80%9A%E9%81%8E-040500856.html'
r2 = requests.get(url2)
web_content2 = r2.text
soup2 = BeautifulSoup(web_content2,'lxml')
soup2
```

```python
articleContent2 = soup2.find_all('p',limit=4)
articleContent2
```

```python
# 將所有tag p中的text取出並放進list。
article2 = []
for p in articleContent2:
    article2.append(p.text)
```

```python
articleAll2 = '\n'.join(article2)
print(articleAll2)
```

```python
d2 = articleAll2.replace('[^\w\s]','').replace('／',"").replace('《','').replace('》','').replace('，','').replace('。','').replace('「','').replace('」','').replace('（','').replace('）','').replace('！','').replace('？','').replace('、','').replace('▲','').replace('…','').replace('：','').replace('?','').replace('(','').replace(')','').replace('【','').replace('】','').replace('/','').replace(';','').replace('；','')
print(d2)
```

```python
import jieba
import nltk
```

```python
jieba.load_userdict('text.txt') # 輸入自己字典的路徑
```

```python
# 設定停用字詞 
stopwords = {}.fromkeys(["也","但","來","個","再","在","的","和","是","有","更","會","可能","有何","從","對","就", '\n','越','為','這種','多','越來',' ','停招','停辦','了','後','有些','或','讓','與','已','於','還是','堅持','1','2','3','4','5','6','7','8','9','0','他','她','個','十','百','千','萬','除了','之外','年','月','日','10','2022','然而','到', '上','下','左','右','左右','而','及','以外','會將','是否','2070','以上','以下','內','外','被','等'])
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence2 = jieba.cut_for_search(d2) 
```

```python
jieba.setLogLevel(20)

Sentence = jieba.cut(d2, cut_all=True)
print('全模式'+": "  + "/ ".join(Sentence) + '\n')   
 
Sentence = jieba.cut(d2, cut_all=False)
print('精確模式'+": " + "/ ".join(Sentence)+ '\n')  
 
Sentence = jieba.cut(d2)  
print('Default為精確模式'+": " + "/ ".join(Sentence)+ '\n')
 
Sentence = jieba.cut_for_search(d2)  
print('搜索引擎模式'+": " + "/ ".join(Sentence)+ '\n')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 文字雲樣式設定
wc = WordCloud(font_path="C:/Windows/Fonts/msjh.ttc", #設置字體
               background_color="white", #背景顏色
               max_words = 35 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
```

```python
# create a python dictionary
hash = {}
for item in Sentence2:

    if item in stopwords:
        continue
    
    if item in hash:
        hash[item] += 1
    else:
        hash[item] = 1
```

```python
hash
```

```python
# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)
```

```python
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(40,20), dpi =1600)
plt.show()
```
