```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#來源：讀+
url6 = 'https://www.readr.tw/post/2913'
r6 = requests.get(url6)
web_content6 = r6.text
soup6 = BeautifulSoup(web_content6,'lxml')
print(soup6.prettify())
```

```python
articleContent6 = soup6.find_all("p",class_="g-article-paragraph")
articleContent6
```

```python
# 將所有tag p中的text取出並放進list。
article6 = []
for p in articleContent6:
    article6.append(p.text)
```

```python
articleAll6 = '\n'.join(article6)
print(articleAll6)
```

```python
d6 = articleAll6.replace('[^\w\s]','').replace('／',"").replace('《','').replace('》','').replace('，','').replace('；','').replace('。','').replace('「','').replace('」','').replace('（','').replace('）','').replace('！','').replace('？','').replace('、','').replace('▲','').replace('…','').replace('：','').replace('?','').replace('(','').replace(')','').replace('【','').replace('】','').replace('/','').replace('-','').replace('20220826 更新華梵大學回應於原文章20220906 新增台灣首府大學回應','').replace('r','').replace('註READr 就報導中提到的學校逐一邀訪將回應及結果一併呈現如下','')
print(d6)
```

```python
import jieba
import nltk
```

```python
jieba.load_userdict('text.txt') # 輸入自己字典的路徑
```

```python
# 設定停用字詞 
stopwords = {}.fromkeys(["也","但","來","個","再","在","的","和","是","有","更","會","可能","有何","從","對","就", '\n','越','為','我們','這種','多','越來',' ','停招','停辦','了','後','有些','或','讓','與','已','於','還是','堅持','1','2','3','4','5','6','7','8','9','0','他','她','個','十','百','千','萬','除了','之外','年','月','日','10','2022','然而','到', '上','下','左','右','左右','而','及','以外','會將','是否','2070','以上','以下','內','外','被','等','尤榮輝','27','都','若','及','上午','下午','嗎','呢','所','之','不','這些','就是','真的','\xa0'])
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence3 = jieba.cut_for_search(d6) 
```

```python
jieba.setLogLevel(20)

Sentence = jieba.cut(d6, cut_all=True)
print('全模式'+": "  + "/ ".join(Sentence) + '\n')   
 
Sentence = jieba.cut(d6, cut_all=False)
print('精確模式'+": " + "/ ".join(Sentence)+ '\n')  
 
Sentence = jieba.cut(d6)  
print('Default為精確模式'+": " + "/ ".join(Sentence)+ '\n')
 
Sentence = jieba.cut_for_search(d6)  
print('搜索引擎模式'+": " + "/ ".join(Sentence)+ '\n')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 文字雲樣式設定
wc = WordCloud(font_path="C:/Windows/Fonts/msjh.ttc", #設置字體
               background_color="white", #背景顏色
               max_words = 35 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
```

```python
# create a python dictionary
hash = {}
for item in Sentence3:

    if item in stopwords:
        continue
    
    if item in hash:
        hash[item] += 1
    else:
        hash[item] = 1
```

```python
hash
```

```python
# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)
```

```python
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(40,20), dpi =1600)
plt.show()
```

```python
wc.to_file('outputby讀+.png')
```

```python
at1Df = pd.DataFrame.from_dict(hash, orient='index', columns = ['詞頻'])
dfi1=at1Df.sort_values(by= ['詞頻'], ascending= False)
dfi1t=dfi1[:35]
print(dfi1t)   
```
