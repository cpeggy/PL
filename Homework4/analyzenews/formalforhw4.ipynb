```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#來源：台灣高等教育產業工會
url4 = 'https://www.theunion.org.tw/2021/09/%E3%80%90%E5%A4%A7%E5%AD%B8%E5%BF%AB%E5%A0%B1%E7%AC%AC313%E6%9C%9F%E3%80%91%E7%A7%81%E6%A0%A1%E9%80%80%E5%A0%B4%E6%96%BC%E5%AD%B8%E7%94%9F%E5%8F%88%E6%9C%89%E4%BD%95%E7%9B%8A%EF%BC%9F/'
r4 = requests.get(url4)
web_content4 = r4.text
soup4 = BeautifulSoup(web_content4,'lxml')
print(soup4.prettify())
```

```python
articleContent4 = soup4.find_all(["p",'/p'])
articleContent4
```

```python
# 將所有tag p中的text取出並放進list。
article4 = []
for p in articleContent4:
    article4.append(p.text)
```

```python
articleAll4 = '\n'.join(article4)
print(articleAll4)
```

```python
d4 = articleAll4.replace('[^\w\s]','').replace('／',"").replace('《','').replace('》','').replace('，','').replace('；','').replace('。','').replace('「','').replace('」','').replace('（','').replace('）','').replace('！','').replace('？','').replace('、','').replace('▲','').replace('…','').replace('：','').replace('?','').replace('(','').replace(')','').replace('【','').replace('】','').replace('/','').replace('台灣高等教育產業工會','').replace('改變高教從加入工會開始','').replace('文謝青龍高教工會大雄分部召集人南華大學通識中心專任教授','').replace('原文取自風傳媒','')
print(d4)
```

```python
import jieba
import nltk
```

```python
jieba.load_userdict('text.txt') # 輸入自己字典的路徑
```

```python
# 設定停用字詞 
stopwords = {}.fromkeys(["也","但","來","個","再","在","的","和","是","有","更","會","可能","有何","從","對","就", '\n','越','為','我們','這種','多','越來',' ','停招','停辦','了','後','有些','或','讓','與','已','於','還是','堅持','1','2','3','4','5','6','7','8','9','0','他','她','個','十','百','千','萬','除了','之外','年','月','日','10','2022','然而','到', '上','下','左','右','左右','而','及','以外','會將','是否','2070','以上','以下','內','外','被','等','尤榮輝','27','都','若','及','上午','下午','嗎','呢','所','之','不','這些','就是','真的','\xa0'])
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence3 = jieba.cut_for_search(d4) 
```

```python
jieba.setLogLevel(20)

Sentence = jieba.cut(d4, cut_all=True)
print('全模式'+": "  + "/ ".join(Sentence) + '\n')   
 
Sentence = jieba.cut(d4, cut_all=False)
print('精確模式'+": " + "/ ".join(Sentence)+ '\n')  
 
Sentence = jieba.cut(d4)  
print('Default為精確模式'+": " + "/ ".join(Sentence)+ '\n')
 
Sentence = jieba.cut_for_search(d4)  
print('搜索引擎模式'+": " + "/ ".join(Sentence)+ '\n')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 文字雲樣式設定
wc = WordCloud(font_path="C:/Windows/Fonts/msjh.ttc", #設置字體
               background_color="white", #背景顏色
               max_words = 35 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
```

```python
# create a python dictionary
hash = {}
for item in Sentence3:

    if item in stopwords:
        continue
    
    if item in hash:
        hash[item] += 1
    else:
        hash[item] = 1
```

```python
hash
```

```python
# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)
```

```python
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(40,20), dpi =1600)
plt.show()
```
