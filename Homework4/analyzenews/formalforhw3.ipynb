```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#來源：遠見
url3 = 'https://www.gvm.com.tw/article/95796'
r3 = requests.get(url3)
web_content3 = r3.text
soup3 = BeautifulSoup(web_content3,'lxml')
print(soup3.prettify())
```

```python
articleContent3 = soup3.find_all('p',limit=6)
articleContent3
```

```python
# 將所有tag p中的text取出並放進list。
article3 = []
for p in articleContent3:
    article3.append(p.text)
```

```python
articleAll3 = '\n'.join(article3)
print(articleAll3)
```

```python
d3 = articleAll3.replace('[^\w\s]','').replace('／',"").replace('《','').replace('》','').replace('，','').replace('。','').replace('「','').replace('」','').replace('（','').replace('）','').replace('！','').replace('？','').replace('、','').replace('▲','').replace('…','').replace('：','').replace('?','').replace('(','').replace(')','').replace('【','').replace('】','').replace('/','').replace(';','').replace('；','').replace('三大絕配組合台式餐點 X 高奢威士忌溫暖人心','').replace('新鮮海產及璀璨夜景 一起來基隆生活吧','')
print(d3)
```

```python
import jieba
import nltk
```

```python
jieba.load_userdict('text.txt') # 輸入自己字典的路徑
```

```python
# 設定停用字詞 
stopwords = {}.fromkeys(["也","但","來","個","再","在","的","和","是","有","更","會","可能","有何","從","對","就", '\n','越','為','這種','多','越來',' ','停招','停辦','了','後','有些','或','讓','與','已','於','還是','堅持','1','2','3','4','5','6','7','8','9','0','他','她','個','十','百','千','萬','除了','之外','年','月','日','10','2022','然而','到', '上','下','左','右','左右','而','及','以外','會將','是否','2070','以上','以下','內','外','被','等','尤榮輝','27','都','若','及','上午','下午'])
```

```python
# 使用cut_for_search(搜尋引擎)斷詞模式並產生字詞頻率的dictionary (並剔除stopwords的計算)
Sentence3 = jieba.cut_for_search(d3) 
```

```python
jieba.setLogLevel(20)

Sentence = jieba.cut(d3, cut_all=True)
print('全模式'+": "  + "/ ".join(Sentence) + '\n')   
 
Sentence = jieba.cut(d3, cut_all=False)
print('精確模式'+": " + "/ ".join(Sentence)+ '\n')  
 
Sentence = jieba.cut(d3)  
print('Default為精確模式'+": " + "/ ".join(Sentence)+ '\n')
 
Sentence = jieba.cut_for_search(d3)  
print('搜索引擎模式'+": " + "/ ".join(Sentence)+ '\n')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 文字雲樣式設定
wc = WordCloud(font_path="C:/Windows/Fonts/msjh.ttc", #設置字體
               background_color="white", #背景顏色
               max_words = 35 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
```

```python
# create a python dictionary
hash = {}
for item in Sentence3:

    if item in stopwords:
        continue
    
    if item in hash:
        hash[item] += 1
    else:
        hash[item] = 1
```

```python
hash
```

```python
# 使用dictionary的內容產生文字雲
wc.generate_from_frequencies(hash)
```

```python
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(40,20), dpi =1600)
plt.show()
```
