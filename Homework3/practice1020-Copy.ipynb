```python
import requests 
import json
import pandas as pd 
import requests as req1
import urllib.request as req
url="https://www.ptt.cc/bbs/AIESEC/index.html"
```

```python
#建立一個 Request物件,附加 Headers 資訊使程式像正常使用者狀態
request=req.Request(url, headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
})
r = req1.get(url)
with req.urlopen(request) as response:
    data=response.read().decode("utf-8")
print(data)
```

```python
import bs4
from bs4 import BeautifulSoup
```

```python
html_doc = data
```

```python
#透過bs4整理且用html.parser解析
root = bs4.BeautifulSoup(r.text,"html.parser")
#找到所有屬性class = "title"
titles = root.find_all("div",class_="title")
#一個一個印出要的資料
temp1=[('標題','網址')]
for title in titles:
    A=title.a.string,"https://www.ptt.cc"+title.a.get("href")
    print(title.a.string,"https://www.ptt.cc"+title.a.get("href"))#取得文章標題)
    temp1.append(A)
```

```python
temp1
```

```python
# 轉換成DataFrame
df1 = pd.DataFrame(temp1)
print(df1)
```

```python
df1
```

```python
#轉csv檔並下載
df1.to_csv('CrawlerResult.csv')
print("csv 下載finish")
```

```python
#將DataFrame轉為dict
temp2 = df1.to_dict()
type(temp2)
temp2
```

```python
#將dict轉json檔並下載
with open('CrawlerResult.json', 'w', encoding='utf-8') as file:
    json.dump(temp2, file, ensure_ascii=False)  

print("json下載finish")
```
